```python
f = open('wine.data', 'r')
import numpy as np
data = []
for line in f:
    row =[]
    line = line.strip('\n')
    for i in line.split(","):
        row.append(i)
    data.append(row)
dataMass = np.array(data)
print(np.shape(dataMass))
f.close()      
```

```python
import pandas as pd
Columns = ["Type","Alcohol","Malic acid","Ash","Alcalinity of as","Magnesium","Total phenols","Flavanoids",
           "Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315","Proline"]
data = pd.DataFrame(data,columns = Columns)
data
```

```python
X = data[["Alcohol","Malic acid","Ash","Alcalinity of as","Magnesium","Total phenols","Flavanoids",
           "Nonflavanoid phenols","Proanthocyanins","Color intensity","Hue","OD280/OD315","Proline"]]
y = data["Type"]
y
import operator
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
result = {}
for i in range(1,51):
    kf = KFold(n_splits=5,shuffle=True,random_state=42)
    neigh = KNeighborsClassifier(n_neighbors=i)
    accuracy = cross_val_score(neigh, X,y, cv=kf)
    result[i]= sum(accuracy)/len(accuracy)
print(max(result.items(), key=operator.itemgetter(1))[0])   
print(result[max(result.items(), key=operator.itemgetter(1))[0]])
```
